```python

df = pd.read_csv(f"results/enron/max_distances.csv")
ms = df.columns.to_list()
df = df.assign(
    **{
        "category": sorted([
            p.split("/")[-1][5:-4] for p in glob(f"results/enron/sigs/*.csv")
        ])
    }
)
msdf = pd.melt(df[ms + ["category"]], id_vars="category", var_name="element", value_name="KL")
display(
    timeline(
        msdf,
        x="category",
        y="KL",
        subcorpus="enron",
        stack=None,
        order=ms,
        name=f"vis_2_enron_timeline",
    )
)
```
